```python
import matplotlib.pyplot as plt
import numpy as np
from style import apply_style, save_figure, COLORS, CATEGORICAL_COLORS, GOLD, TEAL, PURPLE

apply_style()

# Data preparation
categories_organisms = ['C. elegans', 'Fly (Brain)', 'Mouse (Brain)', 'Human (Brain)']
categories_full = categories_organisms + ['H100', 'xAI Colossus (2024)']
x_pos_full = np.arange(len(categories_full))
x_pos_organisms = np.arange(len(categories_organisms))

# --- Plot 1: Emulation Compute Requirements (Time-Based) ---
time_based_compute_mids = np.array([
    (1.04E+09 + 2.69E+09) / 2,
    (2.71E+12 + 4.80E+12) / 2,
    (6.75E+15 + 1.10E+16) / 2,
    (8.50E+18 + 1.39E+19) / 2
])
time_based_compute_errors = np.array([
    (2.69E+09 - 1.04E+09) / 2,
    (4.80E+12 - 2.71E+12) / 2,
    (1.10E+16 - 6.75E+15) / 2,
    (1.39E+19 - 8.50E+18) / 2
])
reference_compute_values = [np.nan, np.nan, np.nan, np.nan, 1E+15, 1E+20]

fig, ax = plt.subplots(figsize=(10, 6))
ax.bar(x_pos_organisms, time_based_compute_mids,
       yerr=time_based_compute_errors,
       align='center', alpha=0.85, ecolor=COLORS['text'], capsize=5, color=GOLD, 
       label='Organisms', edgecolor='white', linewidth=1)
ax.bar(x_pos_full, reference_compute_values,
       align='center', alpha=0.7, color=COLORS['caption'], label='Reference Systems',
       edgecolor='white', linewidth=1)

ax.set_ylabel('Compute Requirements (FLOP/s)')
ax.set_title('Emulation Compute Requirements (Time-Based)')
ax.set_xticks(x_pos_full)
ax.set_xticklabels(categories_full, rotation=30, ha="right")
ax.set_yscale('log')

time_based_yticks_values = [1e8, 1e10, 1e12, 1e14, 1e16, 1e18, 1e20]
time_based_yticks_labels = ['100 MFLOPS', '10 GFLOPS', '1 TFLOPS', '100 TFLOPS', '10 PFLOPS', '1 EFLOPS', '100 EFLOPS']
ax.set_yticks(time_based_yticks_values)
ax.set_yticklabels(time_based_yticks_labels)
ax.set_ylim(1e8, 1e21)
ax.legend(loc='upper left', frameon=True)
ax.grid(True, which="both", linestyle=':', alpha=0.5)

plt.tight_layout()
save_figure(fig, 'emulation-compute-time-based')
plt.show()

# --- Plot 2: Emulation Storage Requirements ---
storage_mids = np.array([
    (1.66E+05 + 3.54E+05) / 2,
    (4.33E+08 + 8.75E+08) / 2,
    (1.08E+12 + 2.17E+12) / 2,
    (1.36E+15 + 2.73E+15) / 2
])
storage_errors = np.array([
    (3.54E+05 - 1.66E+05) / 2,
    (8.75E+08 - 4.33E+08) / 2,
    (2.17E+12 - 1.08E+12) / 2,
    (2.73E+15 - 1.36E+15) / 2
])
reference_storage_values = [np.nan, np.nan, np.nan, np.nan, 8E+10, 1E+16]

fig, ax = plt.subplots(figsize=(10, 6))
ax.bar(x_pos_organisms, storage_mids,
       yerr=storage_errors,
       align='center', alpha=0.85, ecolor=COLORS['text'], capsize=5, color=TEAL, 
       label='Organisms', edgecolor='white', linewidth=1)
ax.bar(x_pos_full, reference_storage_values,
       align='center', alpha=0.7, color=COLORS['caption'], label='Reference Systems',
       edgecolor='white', linewidth=1)

ax.set_ylabel('Storage Requirements (Bytes)')
ax.set_title('Emulation Storage Requirements')
ax.set_xticks(x_pos_full)
ax.set_xticklabels(categories_full, rotation=30, ha="right")
ax.set_yscale('log')

storage_yticks_values = [1e5, 1e7, 1e9, 1e11, 1e13, 1e15]
storage_yticks_labels = ['100 KB', '10 MB', '1 GB', '100 GB', '10 TB', '1 PB']
ax.set_yticks(storage_yticks_values)
ax.set_yticklabels(storage_yticks_labels)
ax.set_ylim(1e5, 1e17)
ax.legend(loc='upper left', frameon=True)
ax.grid(True, which="both", linestyle=':', alpha=0.5)

plt.tight_layout()
save_figure(fig, 'emulation-storage-requirements')
plt.show()

# --- Plot 3: Emulation Compute Requirements (Event-Driven) ---
event_driven_compute_mids = np.array([
    (4.16E+06 + 1.05E+09) / 2,
    (1.08E+10 + 5.11E+11) / 2,
    (2.70E+13 + 3.10E+14) / 2,
    (3.40E+16 + 8.67E+16) / 2
])
event_driven_compute_errors = np.array([
    (1.05E+09 - 4.16E+06) / 2,
    (5.11E+11 - 1.08E+10) / 2,
    (3.10E+14 - 2.70E+13) / 2,
    (8.67E+16 - 3.40E+16) / 2
])

fig, ax = plt.subplots(figsize=(10, 6))
ax.bar(x_pos_organisms, event_driven_compute_mids,
       yerr=event_driven_compute_errors,
       align='center', alpha=0.85, ecolor=COLORS['text'], capsize=5, color=PURPLE, 
       label='Organisms', edgecolor='white', linewidth=1)
ax.bar(x_pos_full, reference_compute_values,
       align='center', alpha=0.7, color=COLORS['caption'], label='Reference Systems',
       edgecolor='white', linewidth=1)

ax.set_ylabel('Compute Requirements (FLOP/s)')
ax.set_title('Emulation Compute Requirements (Event-Driven)')
ax.set_xticks(x_pos_full)
ax.set_xticklabels(categories_full, rotation=30, ha="right")
ax.set_yscale('log')

event_driven_yticks_values = [1e6, 1e8, 1e10, 1e12, 1e14, 1e16, 1e18, 1e20]
event_driven_yticks_labels = ['1 MFLOPS', '100 MFLOPS', '10 GFLOPS', '1 TFLOPS', '100 TFLOPS', '10 PFLOPS', '1 EFLOPS', '100 EFLOPS']
ax.set_yticks(event_driven_yticks_values)
ax.set_yticklabels(event_driven_yticks_labels)
ax.set_ylim(1e6, 1e21)
ax.legend(loc='upper left', frameon=True)
ax.grid(True, which="both", linestyle=':', alpha=0.5)

plt.tight_layout()
save_figure(fig, 'emulation-compute-event-driven')
plt.show()
```
